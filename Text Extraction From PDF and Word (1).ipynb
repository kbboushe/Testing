```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import docx2txt
import PyPDF2

# Text Extraction
from PyPDF2 import PdfReader

def extract_text_from_pdf(pdf_path):
    text = ''
    with open(pdf_path, 'rb') as file:
        reader = PdfReader(file)
        for page in reader.pages:
            text += page.extract_text()
    return text


def extract_text_from_docx(docx_path):
    text = docx2txt.process(docx_path)
    return text

resume_text = extract_text_from_pdf('/Users/rezaboushehri/Desktop/Reza Boushehri_Resume.pdf')
job_description_text = extract_text_from_docx('/Users/rezaboushehri/Desktop/Senior Business Strategy and Program Execution Advisor.docx')

# Vectorization
vectorizer = TfidfVectorizer()
text_vectors = vectorizer.fit_transform([resume_text, job_description_text])

# Similarity Measurement
cosine_sim = cosine_similarity(text_vectors[0:1], text_vectors[1:2])

print(f"Cosine Similarity: {cosine_sim[0][0]}")
```

```python
import fitz  # PyMuPDF
import docx
import spacy
from collections import Counter

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_path):
    text = ''
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text()
    return text

# Function to extract text from a Word document
def extract_text_from_docx(docx_path):
    doc = docx.Document(docx_path)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return '\n'.join(fullText)

# Function to analyze text and extract skills
def extract_skills_from_text(text):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)
    skills_counter = Counter()
    # Define skill entities; customize as needed
    skill_labels = ["NOUN", "PROPN"]
    for ent in doc:
        if ent.pos_ in skill_labels:
            skills_counter[ent.text] += 1
    return skills_counter.most_common()

# Paths to your files
pdf_path = '/Users/rezaboushehri/Desktop/Reza Boushehri_Resume.pdf'
docx_path = '/Users/rezaboushehri/Desktop/Senior Business Strategy and Program Execution Advisor.docx'

# Extract texts
resume_text = extract_text_from_pdf(pdf_path)
job_description_text = extract_text_from_docx(docx_path)

# Extract and display skills from the job description
print("Most Relevant Skills Based on the Job Description:")
job_skills = extract_skills_from_text(job_description_text)
for skill, count in job_skills:
    print(f"- {skill} (Mentions: {count})")

# Optionally, compare these skills with the resume
resume_skills = extract_skills_from_text(resume_text)
print("\nSkills Mentioned in the Resume:")
for skill, count in resume_skills:
    if skill in dict(job_skills):
        print(f"- {skill} (Mentions: {count})")
```

```python
import os
import fitz  # PyMuPDF
import docx
import openai

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text()
    return text

# Function to extract text from DOCX
def extract_text_from_docx(docx_path):
    doc = docx.Document(docx_path)
    return "\n".join([para.text for para in doc.paragraphs])

# Your OpenAI API Key
openai.api_key = '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'  # Ensure this is your actual API key

# Paths
resume_path = '/Users/rezaboushehri/Desktop/Reza Boushehri_Resume.pdf'  # Update this path
job_descriptions_dir = '/Users/rezaboushehri/Desktop/Job Descriptions'

# Extract text from resume
resume_text = extract_text_from_pdf(resume_path)

# Iterate over each file in the job descriptions directory
for filename in os.listdir(job_descriptions_dir):
    full_path = os.path.join(job_descriptions_dir, filename)
    if os.path.isfile(full_path):
        # Determine the file type and extract text
        if filename.lower().endswith('.pdf'):
            job_description_text = extract_text_from_pdf(full_path)
        elif filename.lower().endswith('.docx'):
            job_description_text = extract_text_from_docx(full_path)
        else:
            continue  # Skip files that are not PDF or DOCX

        # Construct the prompt
        prompt_text = f"""
    Provide a ranked list of job postings that best match the analyzed resume/CV, highlighting the matching criteria for transparency and further review by the candidate. Offer the ability to adjust preferences or filters for real-time refinement of the results.
---

        ////////////////////////////////////////////////
        
        1) Given the resume and the separate job descriptions that follow, 
        perform a detailed comparison to determine the most relevant job posting for the candidate.
        
        //////////////////////////////////////////////////////
        
        2) Utilize the specified metrics for analysis: location, job level, 
        years of experience, industry, required and preferred skills, and any stated 
        candidate preferences such as company culture and employment type. 
        
        ///////////////////////////////////////////////////////
        
        3) Provide a ranked list of job postings based on their relevance to the resume, 
        highlighting key matching and mismatching points for each job posting in relation to the resume.
        
        //////////////////////////////////////////////////////
        
        4) Rank the job postings from most to least relevant based on the comparison.
        
        ///////////////////////////////////////
        
        5) Provide a brief explanation for the ranking, noting significant matches or 
        discrepancies between the resume and each job description.
        
        ///////////////////////////////////////////
        \n\nJob Description:\n{job_description_text}\n\nResume:\n{resume_text}"""""

        # Send the request to OpenAI's chat API
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt_text}
            ]
        )

        # Print the response
        print(f"Analysis for {filename}:")
        print(response.choices[0].message['content'])
        print("\n" + "="*50 + "\n")
```
